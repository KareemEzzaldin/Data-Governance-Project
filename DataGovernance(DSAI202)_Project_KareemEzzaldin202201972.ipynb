## Data Cleaning

Dealing with Duplicates

```python
# Check Data Consistency
# Ensure that the data does not have whitespace or case inconsistencies. You can apply string methods to standardize the data.
df = df.applymap(lambda x: x.strip() if isinstance(x, str) else x)
df = df.applymap(lambda x: x.lower() if isinstance(x, str) else x)
```

```python
# Duplicate Removal
df_cleaned = df.drop_duplicates(inplace=True)
df_cleaned
```

```python
data_duplicates = df.duplicated().sum()
print(f"Number of duplicates after cleaning: {data_duplicates}")
```

```python
# get the number of missing data points per column
missing_values_count = df.isnull().sum()
# look at the # of missing points in the first ten columns
missing_values_count[:]
```

```python
# how many total missing values do we have?
total_cells = np.product(df.shape)
total_missing = missing_values_count.sum()

# percent of data that is missing
(total_missing/total_cells) * 100
```

```python
# Filling missing values for numerical and categorical columns
import pandas as pd
# Assuming 'data' is your DataFrame
categorical_columns = ["type", "rating", "director", "cast", "country", "date_added", "duration" ]

# For categorical columns, fill with the mode (most frequent value)
for col in categorical_columns:
    mode_value = df[col].mode()[0]  # Calculate mode
    df[col].fillna(mode_value, inplace=True)  # Fill missing values with mode
```

```python
# Check missing values after applying fill
final_missing = df.isnull().sum()
print("Final missing values by column:\n", final_missing)

#  total number of cells in the DataFrame
total_cells = np.product(df.shape)

# total number of missing values
total_missing = final_missing.sum()

# percentage of missing data
percentage_missing = (total_missing / total_cells) * 100
print("Percentage of missing data:", percentage_missing)
```

# Validation Schema

```python
!pip install pandera
```

```python
import pandera as pa
from pandera import Column, DataFrameSchema, Check, Index
```

```python
netflix_df = pd.read_csv("netflix_titles.csv", encoding="utf-8")
schema = pa.DataFrameSchema(
    {
      "type":Column(str, Check.isin(["Movie", "TV Show"])),
      "title":Column(str),
      "director":Column(str),
      "cast":Column(str),
      "country":Column(str),
      "date_added":Column(str),
      "release_year":Column(int, Check.in_range(1950, 2022), coerce=True),
      "rating":Column(str, Check.isin(["TV-Y", "TV-Y7", "TV-G", "TV-PG", "TV-14", "TV-MA", "G", "PG ", "PG-13", "R", "NC-17"])),
      "duration":Column(str),
      "listed_in":Column(str),
      "description":Column(str),
    },
    unique=["show_id"],
)

validated_df = schema(netflix_df)
print(validated_df)
```

```python
try:
    schema.validate(netflix_df, lazy=True)
except pa.errors.SchemaErrors as exc:
    failure_cases_df = exc.failure_cases
    display(exc.failure_cases)
```

```python
# Schema inference is a function that scans your data and generates a DataFrameSchema with some basic checks;
#  this is intended to provide a starting point for you to tailor and further develop.
!pip install pandera[io]
inferred_schema = pa.infer_schema(netflix_df).to_script()
print(inferred_schema)
```

```python
import string


# define function that operates on entire dataframe
def randomize_values(df):
    for column in df.columns:
        if df[column].dtype == 'O': # check if column has object dtype
            df[column] = [''.join(random.choices(string.ascii_letters + string.digits, k=10))
            for _ in range(len(df))] # generate a list of random strings
    return df
```

## Data Pseudonymization

Scrambling:rearranging or shuffling the characters in a string.

```python
# Define a function to scramble data in a specific column
def scramble_data(data):
    data_list = list(data)
    random.shuffle(data_list)
    return ''.join(data_list)

# Scramble the 'description' column
df['Scrambled_description'] = df['description'].apply(scramble_data)

# Display the result
print(df[['description', 'Scrambled_description']].head())
```

Hashing:converting data into a fixed-size string of bytes using a hash function.

```python
import hashlib
```

```python
# Hash the 'title' column using a lambda function
df['Hashed_title'] = df['title'].apply(lambda x: hashlib.sha256(x.encode()).hexdigest()) # Secure Hash Algorithm 256-bit

# Display the result
df
```

Encryption: Encryption involves transforming data into a ciphertext using an encryption algorithm and a secret key.

```python
from cryptography.fernet import Fernet
```

```python
key = Fernet.generate_key()

# Initialize cipher
cipher = Fernet(key)

# Encrypt the 'country' column using a lambda function
df['Encrypted_country	'] = df['country'].apply(lambda x: cipher.encrypt(x.encode()).decode())
df
```
